```python
import numpy as np
import matplotlib.pyplot as plt

# We will simulate the Law of Large Numbers
# Using a Bernoulli random variable — coin flip
# True mean (p) = 0.5
# We will watch the sample average converge to 0.5 as n grows

p = 0.5          # true probability of heads
n_trials = 10000 # total number of flips

# Simulate all flips at once
np.random.seed(42)
flips = np.random.binomial(1, p, size=n_trials)
print(f"First 20 flips: {flips[:20]}")
print(f"True mean (p): {p}")
print(f"Sample mean after {n_trials} flips: {np.mean(flips):.4f}")
```

```python
# Calculate running average after each flip
# running_avg[i] = average of first i+1 flips
# Watch it start wild and slowly settle at 0.5

running_avg = np.cumsum(flips) / np.arange(1, n_trials + 1)

plt.figure(figsize=(12, 5))

plt.plot(running_avg, color='blue', linewidth=1, label='Sample average')
plt.axhline(y=p, color='red', linewidth=2, linestyle='--', label=f'True mean (p={p})')

plt.xlabel('Number of flips')
plt.ylabel('Sample average')
plt.title('Law of Large Numbers — Sample Average Converging to True Mean')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Run 20 different simulations of 10000 flips each
# Each line = one experiment
# All of them should converge to 0.5
# This shows LLN holds universally - not just for one lucky run

plt.figure(figsize=(12, 5))

n_simulations = 20

for i in range(n_simulations):
    flips_i = np.random.binomial(1, p, size=n_trials)
    running_avg_i = np.cumsum(flips_i) / np.arange(1, n_trials + 1)
    plt.plot(running_avg_i, linewidth=0.8, alpha=0.5)

plt.axhline(y=p, color='red', linewidth=2.5, linestyle='--', label=f'True mean (p={p})')

plt.xlabel('Number of flips')
plt.ylabel('Sample average')
plt.title('LLN — 20 Independent Simulations All Converging to True Mean')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
